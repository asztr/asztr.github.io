```python
import os
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from pylab import rcParams
rcParams['figure.figsize'] = 10, 8
rcParams['figure.dpi'] = 300

import torch
from torch import nn
from torch import distributions
from torch.nn.parameter import Parameter
from torch.utils import data
from tensorboardX import SummaryWriter

from sklearn import cluster, datasets, mixture
from sklearn.preprocessing import StandardScaler

import fastmerl
import coords

def plt_setup(subnum=111, title='', xlabel='', ylabel='', xlim=None, ylim=None):
    plt.subplot(subnum)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    if (xlim is not None):
        plt.xlim(xlim)
    if (ylim is not None):
        plt.ylim(ylim)

def cart_product(*args):
    arglist = [arg if hasattr(arg,'__len__') else [arg] for arg in args]
    return np.array(list(itertools.product(*arglist)))

def nml(vec):
    return np.array(vec)/np.linalg.norm(vec)

def random_uniform_directions(nsamples):
    phi = np.random.uniform(size=nsamples, low=0.0, high=2*np.pi)
    z = np.random.uniform(size=nsamples, low=0.0, high=1)
    x = np.sqrt(1-z**2)*np.cos(phi)
    y = np.sqrt(1-z**2)*np.sin(phi)
    return np.array([x, y, z])

# sigma is standard deviation. I'm using s for variance.
# s = sigma**2
# torch.distributions.Normal and std::normal_distribution use sigma (stddev)
# torch.distributions.MultivariateNormal uses s (variance)

def gauss(x, mu, sig):
    return (1./(sig*np.sqrt(2*np.pi)))*np.exp(-0.5*((x-mu)**2)/sig**2)

def gauss2(x, mu, s):
    return (1./np.sqrt(2*np.pi*s))*np.exp(-0.5*((x-mu)**2)/s)

def gauss2d(x,mu,sig): #only diagonal sigma
    return gauss(x[0],mu[0],sig[0])*gauss(x[1],mu[1],sig[1])

def gauss2d_2(x,mu,s): #only diagonal s
    return gauss2(x[0],mu[0],s[0])*gauss2(x[1],mu[1],s[1])
    #return (1./(np.sqrt(s[0]*s[1])*2*np.pi))*np.exp(-0.5*(((x[0]-mu[0])**2)/s[0] + ((x[1]-mu[1])**2)/s[1]))

def tt(arr):
    return torch.tensor(arr).float()

def npa(arr):
    return np.array(arr)

def to_np(arr):
    return arr.detach().cpu().numpy()

def npextend(*args): #we could add np.tile or similar for non-matching dimensions
    #find shape to match
    shape = None
    for arg in args:
        if hasattr(arg,'__len__'):
            _shape = np.array(arg).shape
            if (shape is not None) and (shape != _shape):
                raise Exception('Multiple arguments with different shapes')
            else:
                shape = _shape
    if (shape is None):
        return args
    else: #tile scalars
        return np.array([arg if hasattr(arg,'__len__') else np.ones(shape)*arg for arg in args])

def set_device(devname):
    if (devname in ["cuda", "cuda:0"]) and (torch.cuda.is_available()):
        dev = torch.device("cuda:0")
        torch.cuda.empty_cache()
        torch.set_default_tensor_type('torch.cuda.FloatTensor')
    elif (devname == "cpu"):
        dev = torch.device('cpu')
        torch.set_default_tensor_type('torch.FloatTensor')
    else:
        print("wrong device name or device not available")
    return dev

dev = set_device("cuda")

theta_max = np.pi/2.
phi_max = 2*np.pi
```

```python
class interp1D():
    def __init__(self, y, x0, xf):
        self.x = np.linspace(x0, xf, len(y))
        self.y = np.array(y)
        self.dx = (xf-x0)/(len(y)-1.)

    def eval_raw(self, x):
        i = ((np.array(x)-self.x[0])/self.dx).astype(np.int32)
        return self.y[i]

    def eval_interp(self, x):
        return np.interp(x, self.x, self.y)

# 2D integral of BRDF(wo) given wi.
def integral_2d_wiwo(theta_i, phi_i, theta_o2, phi_o2, dtheta=1e-2, dphi=1e-2):
    if (theta_o2 == 0.0) or (phi_o2 == 0.0):
        return 0.0

    ntheta, nphi = 1+int(theta_o2/dtheta), 1+int(phi_o2/dphi)
    dxy = dtheta*dphi
    theta_o = np.linspace(0, theta_o2, ntheta)
    phi_o   = np.linspace(0, phi_o2, nphi)

    #grid = np.array(np.meshgrid(theta_i, phi_i, theta_o, phi_o)).reshape(4,-1)
    grid = cart_product(theta_i, phi_i, theta_o, phi_o).T
    jac  = np.sin(grid[2])*np.cos(grid[2])
    
    rangles = coords.ioangles_to_rangles(*grid)
    brdf_mean = merl.eval_raw(*rangles).mean(axis=0)
    integ = np.sum(brdf_mean*jac*dxy)
    return integ

# Compute normalization factor
def compute_norm_factor(delta=5e-3):
    theta_i = np.linspace(0.0, np.pi/2., 100)
    integ = [integral_2d_wiwo(th_i, 0.0, np.pi/2., 2*np.pi, delta, delta) for th_i in theta_i]
    return integ

# 2D CDF evaluated in WIWO
def cdf_2d_wiwo(theta_i, phi_i, theta_o, phi_o, delta=1e-2): #check whether P(theta_o, phi_o) is BRDF or BRDF.cos(theta_o)
    nrm = norm_factor.eval_interp(theta_i) #integral_2d_wiwo(theta_i, phi_i, theta_max, phi_max, delta, delta)
    #mode 1
    u_theta = integral_2d_wiwo(theta_i, phi_i, theta_o, 2*np.pi, delta, delta) / nrm
    if (u_theta > 0.0):
        u_phi = integral_2d_wiwo(theta_i, phi_i, theta_o, phi_o, delta, delta) / (u_theta * nrm)
        return u_theta, u_phi
    #mode 2
    u_phi = integral_2d_wiwo(theta_i, phi_i, np.pi/2., phi_o, delta, delta) / nrm
    if (u_phi <= 0.0):
        return 0.0, 0.0
    u_theta = integral_2d_wiwo(theta_i, phi_i, theta_o, phi_o, delta, delta) / (u_phi * nrm)
    return u_theta, u_phi

merl = fastmerl.Merl('/home/asztrajman/repo/merl/materials/blue-acrylic.fullbin')

# Computation of normalization factor. The value only depends on theta_i. Stored in interp1D.
integ = compute_norm_factor()
norm_factor = interp1D(integ, 0.0, np.pi/2.)
```

```python
## Generation of dataset (800k ~ 10 hrs)
#c++/jit/multicore/cython?

col_names = ['theta_i', 'phi_i', 'theta_o', 'phi_o', 'u_theta', 'u_phi'] #, 'pdf_mean']
Xvars = ['theta_i', 'phi_i', 'u_theta', 'u_phi']
Yvars = ['theta_o', 'phi_o'] #, 'pdf_mean']

def generate_naiverusin_dataset(npts):
    dataset = []
    ioangles = coords.random_directions_naiverusin(size=npts, ret='ioangles')
    ioangles = ioangles.T[(ioangles[0]<=np.pi/2.) & (ioangles[2]<=np.pi/2.)] #npts is estimative
    for ioangle in ioangles:
        u_theta, u_phi = cdf_2d_wiwo(*ioangle)
        pdf_mean = merl.eval_raw_wiwo(*ioangle).mean()/norm_factor.eval_interp(ioangle[0])
        dataset += [[*ioangle, u_theta, u_phi, pdf_mean]]
    return np.array(dataset)

def generate_uniform_dataset(npts):
    wi = coords.random_directions(npts)
    wo = coords.random_directions(npts)
    ioangles = np.array([*wi, *wo]).T
    dataset = []
    for ioangle in ioangles:
        u_theta, u_phi = cdf_2d_wiwo(*ioangle)
        pdf_mean = merl.eval_raw_wiwo(*ioangle).mean()/norm_factor.eval_interp(ioangle[0])
        dataset += [[*ioangle, u_theta, u_phi, pdf_mean]]
    return np.array(dataset)

def generate_naiveuniform_dataset(npts):
    wi = np.random.uniform(low=[0,0], high=[theta_max, phi_max], size=(npts,2)).T
    wo = np.random.uniform(low=[0,0], high=[theta_max, phi_max], size=(npts,2)).T
    ioangles = np.array([*wi, *wo]).T
    dataset = []
    for ioangle in ioangles:
        u_theta, u_phi = cdf_2d_wiwo(*ioangle)
        pdf_mean = merl.eval_raw_wiwo(*ioangle).mean()/norm_factor.eval_interp(ioangle[0])
        dataset += [[*ioangle, u_theta, u_phi, pdf_mean]]
    return np.array(dataset)

dataset = generate_naiveuniform_dataset(600000)
#dataset = np.load('dataset_l2l_200k.npy')
print(dataset.shape)
# np.save('dataset_l2l_200k_uniform.npy', dataset)

# df = pd.DataFrame(dataset, columns=col_names)
# datasetX = df[Xvars].values
# datasetY = df[Yvars].values
```
